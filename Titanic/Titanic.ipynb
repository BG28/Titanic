```python
avgAgeMr=titanic[titanic['Name'].str.contains('Mr. ')]['Age'].mean()
avgAgeMrs=titanic[titanic['Name'].str.contains('Mrs. ')]['Age'].mean()
avgAgeMiss=titanic[titanic['Name'].str.contains('Miss. ')]['Age'].mean()
avgAgeMas=titanic[titanic['Name'].str.contains('Master. ')]['Age'].mean()
```

```python
titanic['Title'] = titanic['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
 	'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Others')
```

```python
titanic['Title'].value_counts()
```

```python
titanic['Title']=titanic['Title'].replace('Mlle','Miss')
```

```python
mr=titanic[titanic['Title']=='Mr'].fillna(value=avgAgeMr)
titanic.update(mr)
```

```python
miss=titanic[titanic['Title']=='Miss'].fillna(value=avgAgeMiss)
titanic.update(miss)
```

```python
mas=titanic[titanic['Title']=='Master'].fillna(value=avgAgeMas)
titanic.update(mas)
```

```python
mrs=titanic[titanic['Title']=='Mrs'].fillna(value=avgAgeMrs)
titanic.update(mrs)
```

```python
def family_size(number):
    if number==1:
        return 'Alone'
    elif number>1 and number<5:
        return 'Small'
    else:
        return 'Large'
```

```python
titanic['family_size']=titanic['Family'].apply(family_size)
```
